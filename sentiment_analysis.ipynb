```python
import nltk                                                   #to download lexicon data for polarity score
import tweepy                                                 #to access twitter API
import pandas as pd                                           #data manipulation
from nltk.sentiment.vader import SentimentIntensityAnalyzer   #Sentiment analyzer and lexicon
```

```python
#twitter API data for authentication
consumer_key = ""
consumer_secret_key = ""

access_token = ""
access_token_secret = ""

```

```python
#authentication using OAuth
auth = tweepy.OAuthHandler(consumer_key,consumer_secret_key)    #provide reqd info for API authentication
auth.set_access_token(access_token,access_token_secret)

api = tweepy.API(auth)
```

```python
tweets = api.search("Machine learning",count=100)                   #provide the search term with the reqd no. of data,i.e, count
data = pd.DataFrame(data=[tweet.text for tweet in tweets],columns=["tweets"])      #convert the tweets to pandas dataframe

data.head(5)
```

```python
nltk.download(["vader_lexicon"])                 #download lexicon data 
polarity_list = []                               #list to store sentiment score
sentiment_intensity = SentimentIntensityAnalyzer()
for index,row in data.iterrows():
    sentiment_polarity = sentiment_intensity.polarity_scores(row["tweets"])
    polarity_list.append(sentiment_polarity)    #append the polarity score to a list
    
polarity_series = pd.Series(polarity_list)      #convert polarity list to pandas series
 
```

```python
data["polarity"] = polarity_series.values    #store the the polarity value to a new column in data - "polarity"

display(data.head(100))
```
